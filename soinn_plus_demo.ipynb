```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from itertools import compress
from numpy.random import normal
from soinn.soinnplus import SoinnPlus
```

```python
# generate data
n = 1000
sigma = 0.07
c = 10 * np.random.rand(n) - 5
X = np.array([[normal(c[i], sigma), normal(np.sin(c[i]), sigma)] for i in range(len(c))])
```

```python
# initialize SOINN
s = SoinnPlus(delete_node_period=300, init_node_num=3)
s.fit(X)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
plt.scatter(s.nodes[:,0], s.nodes[:,1])
plt.show()
```

```python
from scipy.sparse import triu
for x, y in triu(s.adjacent_mat).todok().items():
    print(x, y)
```

```python
# show SOINN's state
# plt.plot(X[:, 0], X[:, 1], 'cx')
nodes = s.nodes
for k in s.adjacent_mat.keys():
    plt.plot(nodes[k, 0], nodes[k, 1], 'k')
plt.plot(nodes[:, 0], nodes[:, 1], 'ro')
plt.grid(True)
```
